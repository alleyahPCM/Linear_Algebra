2. Without the use of any python libraries or modules, develop a function that can multiply two numpy matrices.

```python
def matrix_multiplication(matrix_a, matrix_b):
    if len(matrix_a[0]) != len(matrix_b):
        raise ValueError("Matrices cannot be multiplied")

    result = [[0 for _ in range(len(matrix_b[0]))] for _ in range(len(matrix_a))]

    for i in range(len(matrix_a)):
        for j in range(len(matrix_b[0])):
            for k in range(len(matrix_b)):
                result[i][j] += matrix_a[i][k] * matrix_b[k][j]

    return result

matrix_a = [
    [1, 3],
    [2, -1]
]
matrix_b = [
    [2, 0, -4],
    [5, -2, 6]
]

# Multiply matrices
result = matrix_multiplication(matrix_a, matrix_b)
if result:
    print("Matrix A:")
    for row in matrix_a:
        print(row)
    print("\nMatrix B:")
    for row in matrix_b:
        print(row)
    print("\nResult:")
    for row in result:
        print(row)
```

3. The rule of distributivity states that given two matrices $A$ and $B$ and a scalar, $k$, then $k(A+B)=kA+kB$. Instead of writing a proof mathematically, develop two codes for $k(A+B)$ and $kA+kB$.

```python
def matrix_addition(matrix_a, matrix_b):
    if len(matrix_a) != len(matrix_b) or len(matrix_a[0]) != len(matrix_b[0]):
        return "Matrices must have the same dimensions"
    
    result = []
    for i in range(len(matrix_a)):
        row = []
        for j in range(len(matrix_a[0])):
            row.append(matrix_a[i][j] + matrix_b[i][j])
        result.append(row)
    
    return result

def scalar_multiplication(matrix, k):
    result = []
    for i in range(len(matrix)):
        row = []
        for j in range(len(matrix[0])):
            row.append(k * matrix[i][j])
        result.append(row)
    
    return result

matrix_a = [
    [2, 3],
    [4, 5]
]

matrix_b = [
    [1, 0],
    [2, 1]
]

k = 3

# k(A + B)
k_times_A_plus_B = scalar_multiply(matrix_addition(matrix_a, matrix_b), k)

# kA + kB
kA_plus_kB = matrix_addition(scalar_multiplication(matrix_a, k), scalar_multiplication(matrix_b, k))

if k_times_A_plus_B and kA_plus_kB:
    print("k =", k)
    print("\nMatrix A:")
    for row in matrix_a:
        print(row)
    print("\nMatrix B:")
    for row in matrix_b:
        print(row)
    print("\nk(A + B):")
    for row in k_times_A_plus_B:
        print(row)
    print("\nkA + kB:")
    for row in kA_plus_kB:
        print(row)
```

4. Without using a python library or modules develop a function that can extract the diagonal of a numpy matrix.

```python
def extract_diagonal(matrix):
    if not matrix:
        return []

    num_rows = len(matrix)
    num_cols = len(matrix[0]) if matrix[0] else 0

    if num_rows != num_cols:
        return []

    diagonal = []
    for i in range(num_rows):
        for j in range(num_cols):
            if i == j:
                diagonal.append(matrix[i][j])

    return diagonal

matrix = [
    [1, 5, 8, 10],
    [11, 2, 6, 9],
    [14, 12, 3, 7],
    [16, 15, 13, 4]
]

result = extract_diagonal(matrix)

if result:
    print("Matrix:")
    for row in matrix:
        print(row)
    print("\nDiagonal of the matrix:")
    print(result)
```

5. Without using a python library or modules develop a function that can find a trace of a numpy matrix.

```python
def matrix_trace(matrix):
    if len(matrix) != len(matrix[0]):
        return "Matrix must have the same dimensions"
    
    trace = 0
    for i in range(len(matrix)):
        trace += matrix[i][i] 
    
    return trace

matrix = [
    [1, 5, 8, 10],
    [11, 2, 6, 9],
    [14, 12, 3, 7],
    [16, 15, 13, 4]
]

print("Trace =", matrix_trace(matrix))
```
